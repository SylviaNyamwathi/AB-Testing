 # **Fast Food Marketing Campaign A\B Test Analysis**

### This is an analysis of [Fast Food Marketing Campaign A\B Test dataset](https://www.kaggle.com/datasets/chebotinaa/fast-food-marketing-campaign-ab-test).

# **Goal of the Test**
The main goal of this A/B test was to test which of the three campaigns resulted in the highest sales.

- **Null Hypothesis**: There is no statistically significant difference between the campaigns in terms of generated sales. In other words, they are equally successful.

- **Alternate Hypothesis**: There is a statistically significant difference between the campaigns in terms of generated sales. In other words, one campaign performs better than the other.

For the analysis of A/B test results, we will use a confidence level of 99% because, as we are comparing multiple campaigns in pairs, we suffer from the multiple testing problem.

# **Target Metric**
The primary metric for this analysis will be the total sales generated by each campaign.

# **Calculations t-test**
First, we are aggregating sales of all four weeks per Location_ID.  The resulting table has 137 rows, which is in alignment with the above data validation queries, telling us we have 137 distinct stores in our sample.



Second, we have to conduct two-tailed test. Two-tailed test in contrast to a one-tailed t-test, is appropriate if you want to determine if there is any difference between groups you are comparing, which is our case.

### t-test **Campaign 1** vs. **Campaign 2**



The p-value is 0.00128, which is significantly less than 0.01 (99% confidence level). This means we reject the null hypothesis and conclude that the difference between **Campaign 1** and **Campaign 2** is statistically significant, indicating that **Campaign 1** performs better than **Campaign 2**.


### t-test **Campaign 2** vs. **Campaign 3**



The p-value is 0.0136, which is greater than 0.01, so we fail to reject the null hypothesis. This implies that the difference between **Campaign 2** and **Campaign 3** is not statistically significant, meaning there is no evidence to suggest that one campaign performs better than the other.

### t-test **Campaign 1** vs. **Campaign 3**



The p-value is 0.43, which is much greater than 0.01. We also fail to reject the null hypothesis in this case, indicating that there is no significant difference in performance between **Campaign 1** and **Campaign 3**.
